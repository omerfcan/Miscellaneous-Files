Data Information:

Credit score cards are a common risk control method in the financial industry. It uses personal information and data submitted by credit card applicants to predict the probability of future defaults and credit card borrowings. The bank is able to decide whether to issue a credit card to the applicant. Credit scores can objectively quantify the magnitude of risk.

```python
df = pd.read_csv('application_record.csv')
```

Let's see the scatter plot of the two columns which are "days birth" and "days employed"

```python
df2=df.copy()
df2.dropna(axis=0,subset=['OCCUPATION_TYPE'], inplace=True)
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df2, x=(df2["DAYS_BIRTH"]*(-1)), y=(df2["DAYS_EMPLOYED"]*(-1)), alpha=0.1,lw=0)
```

Let's see some other graphics

```python
df30=df.copy()
```

```python
df30["Age in Years"] = (df30["DAYS_BIRTH"] / -365)
```

```python
sns.histplot(data = df, x=df30["Age in Years"], linewidth=2,edgecolor='black',
             color='red',bins=45,alpha=0.4)
plt.xlabel("Age in Years")
```

```python
df3 = df.copy()
```

```python
df3.sort_values(by=["AMT_INCOME_TOTAL"], inplace=True)
```

```python
df4 = df3[df3["AMT_INCOME_TOTAL"] <= 160000]
```

```python
df4["NAME_FAMILY_STATUS"].unique()
```

```python
plt.figure(figsize=(14,6))
sns.boxplot(x="NAME_FAMILY_STATUS",y = "AMT_INCOME_TOTAL", hue="FLAG_OWN_REALTY", data=df4,order=['Married', 'Civil marriage', 'Widow', 'Single / not married', 'Separated',] )
plt.legend(bbox_to_anchor=(1,0.5))
```
